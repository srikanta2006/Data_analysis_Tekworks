### 🎯 Your Task:
Using the existing DataFrame, perform the following operations:

✅ Print the details (name, age, marks) of the student who scored the highest marks.

✅ Display the details of students who are 22 years old or older.

✅ Calculate and print the average marks of all students after the bonus.


```python
names = pd.Series(['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Rahul'])
ages = pd.Series([20, 21, 19, 22, 20, 21])
marks = pd.Series([85, 92, 78, 95, 88, 89])
data = {'Name': names, 'Age': ages, 'Marks': marks}
student_df = pd.DataFrame(data)
```

```python
highest_marks_student = student_df.loc[student_df['Marks'].idxmax()]
print("Details of the student with the highest marks:")
display(highest_marks_student)
```

```python
student_df.loc["Tharushi"] = [23, 88]
display(student_df)
```

```python
print(student_df.loc[student_df['Age'] >= 22])
```

```python
student_df.loc[student_df['Marks'] < 80, 'Marks'] += 10
display(student_df)
```

```python
average_marks = student_df['Marks'].mean()
print(f"The average marks of all students after the bonus is: {average_marks:.2f}")
```

Your manager wants you to clean the data, fix inconsistencies, derive business metrics, and explore relationships. The cleaned data should enable deeper dashboards later.

```python
path = '/content/drive/My Drive/Data_sets/OnlineRetail.csv'
df=pd.read_csv(path, encoding='latin-1')
```

```python
print("DataFrame Head:")
print(df.head())
print("\nDataFrame Tail:")
print(df.tail())
print("\nDataFrame Info:")
print(df.info())
print("\nDataFrame Shape:")
x=df.shape
print(x)
print("\nDataFrame Data Types:")
print(df.dtypes)
```

```python
#Missing data handling
print(df.isnull().sum())
```

```python
#droping rows with null customer ID
df.dropna(subset=['CustomerID'], inplace=True)
```

```python
#filling missing description with Unknown
df['Description']=df['Description'].fillna("Unknown")
```

```python
#formating date column
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'], format='mixed')
```

### Detect and handle bad data

```python
# Detect transactions with negative quantities
negative_quantity_transactions = df[df['Quantity'] < 0]
print("Transactions with negative quantity (returns):")
display(negative_quantity_transactions.head())
print(f"\nNumber of transactions with negative quantity: {len(negative_quantity_transactions)}")

# Decide how to handle negative quantities - for this task, let's drop them
df = df[df['Quantity'] >= 0]
print(f"\nDataFrame shape after dropping negative quantities: {df.shape}")
```

```python
# Detect transactions with zero or negative unit prices
bad_price_transactions = df[df['UnitPrice'] <= 0]
print("\nTransactions with zero or negative unit price:")
display(bad_price_transactions.head())
print(f"\nNumber of transactions with zero or negative unit price: {len(bad_price_transactions)}")

# Decide how to handle zero or negative unit prices - let's replace them with the mean
mean_unit_price = df[df['UnitPrice'] > 0]['UnitPrice'].mean()
df.loc[df['UnitPrice'] <= 0, 'UnitPrice'] = mean_unit_price

print(f"\nDataFrame shape after adjusting bad prices: {df.shape}")
```

```python
print(f"duplicate entries: {df.duplicated().sum()}")
```

```python
df.drop_duplicates(inplace=True)
print(df.describe())
print(f"shape of the data set after all changes {df.shape}")
```

```python
#Filtering & Sorting
#Quantity > 1000
high_quantity_transactions=df.loc[df['Quantity']>1000]
print(high_quantity_transactions.head())
print(f"Number of transactions with quantity > 1000: {len(high_quantity_transactions)}")
```

```python
#highest revenue orders(sorting )-descending
#revenue=UnitPrice * Quantity
df['Revenue']=df['Quantity']*df['UnitPrice']
sorted_df=df.sort_values(by='Revenue', ascending=False)
display(sorted_df.head())
```

```python
# Create a small custom DataFrame product_info
product_info = pd.DataFrame({
    'StockCode': ['85123A', '71053', '84406B', '84029G', '22613'],
    'Category': ['Home Decor', 'Home Decor', 'Home Decor', 'Seasonal', 'Party Supplies'],
    'Supplier': ['Supplier A', 'Supplier B', 'Supplier C', 'Supplier A', 'Supplier D']
})

print("Custom DataFrame 'product_info':")
display(product_info)
```

```python
# Merge df with product_info on 'StockCode'
merged_df = pd.merge(df, product_info, on='StockCode', how='left')

print("Merged DataFrame:")
display(merged_df.head())
print(f"\nShape of the merged DataFrame: {merged_df.shape}")
```

```python
#Group by Country and compute total revenue, mean revenue per order, and count of orders
country_stats = merged_df.groupby('Country').agg({
    'Revenue': ['sum', 'mean'],
    'InvoiceNo': 'count'
})
country_stats.columns = ['Total_Revenue', 'Mean_Revenue_Per_Order', 'Count_of_Orders']
print(country_stats)
```

```python
#Rename StockCode to Product_ID, UnitPrice to Price
merged_df.rename(columns={'StockCode':'Product_ID', 'UnitPrice':'Price'})
```

```python
# Use .str methods on Description
# Convert Description to lowercase for case-insensitive searching
merged_df['Description_lower'] = merged_df['Description'].str.lower()

# Filter products whose description contains "glass"
glassware_df = merged_df[merged_df['Description_lower'].str.contains('glass', na=False)]
print("Products with 'glass' in their description:")
display(glassware_df.head())

# Filter products whose description contains "christmas" or "xmas"
christmas_products = merged_df[merged_df['Description_lower'].str.contains('christmas|xmas', na=False)]
print("\nProducts with 'christmas' or 'xmas' in their description:")
display(christmas_products.head())
```
