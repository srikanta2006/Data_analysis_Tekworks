## 3. Categorical Plots (Why: compare groups/categories)

- **Barplot** (`sns.barplot`): shows aggregate (mean) for categories.
- **Countplot** (`sns.countplot`): shows counts of categorical values.
- **Boxplot** (`sns.boxplot`): shows median, quartiles, and outliers.
- **Violinplot** (`sns.violinplot`): boxplot + KDE.

Use `hue` to split by another categorical variable.



```python
# Bar Plot - Average Tip by Gender
sns.barplot(x='sex', y='tip', data=tips)
plt.title('Bar Plot - Average Tip by Gender')
plt.show()
```

```python
# Count Plot - Frequency of Days
sns.countplot(x='day', data=tips)
plt.title('Count Plot - Frequency of Days')
plt.show()
```

```python
# Box Plot - Total Bill by Day
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot - Total Bill by Day')
plt.show()
```

```python
# Violin Plot - Distribution + Summary
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot - Total Bill by Day')
plt.show()
```

## 4. Relationship Plots (Why: see how two numeric variables relate)
**Scatterplot (sns.scatterplot)**: point cloud to inspect correlation.

**Lineplot (sns.lineplot)**: trends over an ordered variable.

**Relplot (sns.relplot)**: figure-level interface for relational plots; supports faceting.

```python
# Scatter Plot - Total Bill vs Tip
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.title('Scatter Plot - Total Bill vs Tip')
plt.show()
```

```python
# Line Plot - Tip vs Party Size (useful for trend-like visuals)
sns.lineplot(x='size', y='tip', data=tips)
plt.title('Line Plot - Tip vs Party Size')
plt.show()
```

```python
# Relplot with Facets
sns.relplot(x='total_bill', y='tip', hue='sex', col='day', data=tips)
plt.suptitle('Relplot - Tip vs Total Bill by Day & Gender', y=1.02)
plt.show()
```

##Matrix & Correlation Plots (Why: inspect relationships between many variables)
**Heatmap (sns.heatmap)**: visualize correlation matrix or any 2D numerical table.

**Clustermap (sns.clustermap)**: heatmap + hierarchical clustering to find groups.

```python
# Select only numeric columns
iris_num = iris.select_dtypes(include=['float', 'int'])

# Compute correlation
corr = iris_num.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True,cmap='coolwarm')
plt.title("Heatmap - Correlation of Iris Features")
plt.show()
```

```python
# Select only numeric columns (exclude 'species')
iris_num = iris.select_dtypes(include=['float', 'int'])

# Compute correlation matrix
corr = iris_num.corr()

# Create the clustermap
sns.clustermap(corr, cmap='coolwarm', annot=True)
plt.suptitle('Clustermap - Grouped Correlations (Numeric Only)', y=1.02)
plt.show()
```

```python
Q1 = new_df['Item_price'].quantile(0.25)
Q3 = new_df['Item_price'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
new_df_cleaned = new_df[(new_df['Item_price'] >= lower_bound) & (new_df['Item_price'] <= upper_bound)].copy()

print(f"Original shape: {new_df.shape}")
print(f"Shape after outlier removal: {new_df_cleaned.shape}")
```

## Time series analysis

### Subtask:
Analyze the number of issues reported over time using the 'Issue_reported at' column. Aggregate issues by day or week and plot the trend.


**Reasoning**:
Resample the data by day and week and plot the trends.



```python
daily_issues = new_df_cleaned.set_index('Issue_reported at').resample('D').size()
plt.figure(figsize=(12, 6))
daily_issues.plot()
plt.title('Daily Issue Reporting Trend')
plt.xlabel('Date')
plt.ylabel('Number of Issues')
plt.show()

weekly_issues = new_df_cleaned.set_index('Issue_reported at').resample('W').size()
plt.figure(figsize=(12, 6))
weekly_issues.plot()
plt.title('Weekly Issue Reporting Trend')
plt.xlabel('Date')
plt.ylabel('Number of Issues')
plt.show()
```

## Time series analysis

### Subtask:
Analyze how 'CSAT Score' changes over time. Aggregate CSAT scores by day or week and plot the trend.


**Reasoning**:
Set 'Issue_reported at' as index, resample by day and week, calculate mean CSAT score, and plot the daily and weekly trends as requested.



```python
new_df_cleaned.set_index('Issue_reported at', inplace=True)

daily_csat = new_df_cleaned['CSAT Score'].resample('D').mean()
plt.figure(figsize=(12, 6))
daily_csat.plot()
plt.title('Daily CSAT Score Trend')
plt.xlabel('Date')
plt.ylabel('Average CSAT Score')
plt.show()

weekly_csat = new_df_cleaned['CSAT Score'].resample('W').mean()
plt.figure(figsize=(12, 6))
weekly_csat.plot()
plt.title('Weekly CSAT Score Trend')
plt.xlabel('Date')
plt.ylabel('Average CSAT Score')
plt.show()
```

## Time series analysis

### Subtask:
Explore other time-based patterns related to handling time or specific categories if relevant.


**Reasoning**:
Resample the connected_handling_time column by day and week, calculate the mean, and plot the daily and weekly trends.



```python
daily_handling_time = new_df_cleaned['connected_handling_time'].resample('D').mean()
plt.figure(figsize=(12, 6))
daily_handling_time.plot()
plt.title('Daily Average Connected Handling Time')
plt.xlabel('Date')
plt.ylabel('Average Handling Time')
plt.show()

weekly_handling_time = new_df_cleaned['connected_handling_time'].resample('W').mean()
plt.figure(figsize=(12, 6))
weekly_handling_time.plot()
plt.title('Weekly Average Connected Handling Time')
plt.xlabel('Date')
plt.ylabel('Average Handling Time')
plt.show()
```
