```python
farm_a = np.array([
    [10, 12, 11],
    [9, 11, 10],
    [12, 13, 14],
    [8, 7, 9]
])


farm_b = np.array([
    [20, 19, 21],
    [18, 20, 19],
    [22, 23, 21],
    [17, 18, 16]
])


farm_c = np.array([
    [15, 16, 14],
    [14, 13, 15],
    [16, 17, 18],
    [12, 11, 13]
])
```

```python
production_all = np.stack([farm_a, farm_b, farm_c])
print("\nStacked production data for all farms:")
print(production_all)
print("\nShape of the stacked array:")
print(production_all.shape)
print("\nSize of the stacked array (total elements):")
print(production_all.size)
print("\nData type of the stacked array:")
print(production_all.dtype)
print("\nNumber of dimensions of the stacked array:")
print(production_all.ndim)
```

```python
# 1. Per-farm totals (arithmetic & aggregation)
print("1. Per-Farm Aggregations ")
total_per_farm = np.sum(production_all, axis=(1, 2))
mean_per_farm = np.mean(production_all, axis=(1, 2))
std_per_farm = np.std(production_all, axis=(1, 2))
var_per_farm = np.var(production_all, axis=(1, 2))

print(f"Total production per farm: {total_per_farm}")
print(f"Mean production per farm: {mean_per_farm}")
print(f"Standard deviation per farm: {std_per_farm}")
print(f"Variance per farm: {var_per_farm}\n")


# 2. Per-hour matrix multiplication
print("--- 2. Weighted Energy Calculation (Matrix Multiplication) ---")
hour_weights = np.array([0.5, 1.0, 0.75])
# Using @ operator for matrix multiplication (dot product)
weighted_per_panel = np.dot(production_all ,hour_weights)
print(f"Weighted energy per panel (shape: {weighted_per_panel.shape}):")
print(weighted_per_panel, "\n")


# 3. Broadcasting adjustments
print("--- 3. Broadcasting Calibration Offset ---")
# The operator notices a fixed calibration offset of +0.5 kWh for all panels in Farm B.
# production_all[1] selects all data for Farm B.
production_all[1] = production_all[1] + 0.5
print("Farm B data after adding 0.5 kWh offset:")
print(production_all[1], "\n")


# 4. Element-wise arithmetic (Z-score normalization)
print("--- 4. Z-Score Normalization ---")
# To broadcast correctly, we need to reshape the (3,) stats arrays to (3, 1, 1)
mean_reshaped = mean_per_farm.reshape(3, 1, 1)
std_reshaped = std_per_farm.reshape(3, 1, 1)
normalized_data = (production_all - mean_reshaped) / std_reshaped
print(f"Normalized data (Z-scores) (shape: {normalized_data.shape}):")
# Showing first farm's normalized data for brevity
print("Farm A (normalized):")
print(normalized_data[0], "\n")


# 5. Applying Universal Functions (ufuncs)
print("--- 5. Universal Function Demonstrations ---")
# Exponential and Logarithmic
print("Exponential (first row of Farm A's normalized data):", np.exp(normalized_data[0, 0, :]))
# Trigonometric
print("Sine (first row of Farm A's normalized data):", np.sin(normalized_data[0, 0, :]))
# Rounding
print("Ceiling of Farm B's adjusted data:", np.ceil(production_all[1]))
# Comparison and Logical
high_output_mask = production_all > 20
print(f"\nNumber of high output instances (>20 kWh): {np.sum(high_output_mask)}")
# Create another mask, e.g., for Farm B
farm_b_mask = np.zeros_like(production_all, dtype=bool)
farm_b_mask[1, :, :] = True
# Combine masks
high_output_in_farm_b = np.logical_and(high_output_mask, farm_b_mask)
print(f"Number of high output instances in Farm B: {np.sum(high_output_in_farm_b)}\n")


# 6. Broadcast-based summary table
print("--- 6. Broadcast-Based Summary Table ---")
panel_totals_per_farm = np.sum(production_all, axis=2)
print(f"Total production per panel for each farm (shape: {panel_totals_per_farm.shape}):")
print(panel_totals_per_farm)
# A panel needs maintenance if its total production is less than 40
needs_maintenance = np.any(panel_totals_per_farm < 40, axis=1)
print(f"\nNeeds maintenance (any panel total < 40): {needs_maintenance}\n")


# 7. Final KPIs (Key Performance Indicators)
print("--- 7. Final KPIs ---")
overall_total = np.sum(production_all)
print(f"Overall total production from all farms: {overall_total:.2f} kWh")

# Find the peak production value and its location
peak_value = np.max(production_all)
peak_flat_index = np.argmax(production_all)
peak_farm, peak_panel, peak_hour = np.unravel_index(peak_flat_index, production_all.shape)
print(f"Peak production of {peak_value:.2f} kWh was at Farm {peak_farm}, Panel {peak_panel}, Hour {peak_hour}")

# Average production per panel (averaging the panel totals)
average_per_panel = np.mean(panel_totals_per_farm)
print(f"Average production per panel across all farms: {average_per_panel:.2f} kWh")

```
